```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt

import seaborn as sns
```

```python
coins = pd.read_csv("coins.csv")
```

```python
sns.scatterplot(x="reflectance",y="weight",data=coins, hue="denomination")
```

```python
rupee1 = zip(coins[coins.denomination==1].reflectance.to_list(), coins[coins.denomination==1].weight.to_list())
rupee2 = zip(coins[coins.denomination==2].reflectance.to_list(), coins[coins.denomination==2].weight.to_list())
rupee1,rupee2 = list(rupee1),list(rupee2)
```

```python
def error(a,b,k,datapoint):
    return np.dot([a,b],datapoint) - k
    
print(error(2,3,4,rupee1[0]))
```
